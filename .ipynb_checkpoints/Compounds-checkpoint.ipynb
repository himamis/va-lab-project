```python
import mysql.connector as con

import pandas as pd  # for loading data

#plots the figures in place instead of a new window
%matplotlib inline  
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA

import numpy as np

from sklearn import cluster

import matplotlib.pyplot as plt
```

```python
connection = con.connect(user = 'root', password = "root", database = 'foodb')
```

```python
compounds_all = pd.read_sql("SELECT * FROM compounds", connection)
```

```python
compounds_all.columns.values
```

```python
#compounds = compounds_all
numeric_columns = ['moldb_alogps_logp','moldb_alogps_logs', 'moldb_average_mass', 'moldb_pka', 'moldb_logp']
for col in numeric_columns:
    compounds[[col]] = compounds[[col]].apply(pd.to_numeric)
for col in numeric_columns:
    compounds = compounds[np.isfinite(compounds[col])]
```

```python
compounds[numeric_columns].describe()
```

```python
pd.scatter_matrix(compounds[numeric_columns], alpha=0.2, figsize=(10, 10), diagonal='kde')
```

```python
compounds[numeric_columns].plot()
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(compounds[numeric_columns])
```

```python
print(pca.explained_variance_ratio_) 
```

```python
pca_trans = pca.transform(compounds[numeric_columns])
```

```python
plt.scatter(x=pca_trans[:, 0], y=pca_trans[:, 1])
```

```python
k_means = cluster.KMeans(n_clusters=2)
```

```python
k_means.fit(compounds[numeric_columns])
```

```python
kmeans_trans = k_means.transform(compounds[numeric_columns])
```

```python
plt.scatter(x=kmeans_trans[:,0], y=kmeans_trans[:,1])
```
